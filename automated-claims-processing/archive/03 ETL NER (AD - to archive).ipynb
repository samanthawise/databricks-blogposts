# Extract Entities
- Parse JSON in NER column to extract name, birth_date and policy_number

```python
%run "./resources/init" 
```

```python
from pyspark.sql.functions import col, from_json
from pyspark.sql.types import StructType, StructField, StringType
```

```python
%python

#define the path to the gold table

df = spark.table(f"{CATALOG}.{SCHEMA}.analysis_gold")
```

```python
# Select necessary columns along with NER fields
df_extracted = df.select(
    col("path"),
    col("modificationTime"),
    col("audio_duration"),
    col("sentiment"),
    col("summary"),
    col("classification"),
    col("email_response"),
    col("ner.name").alias("name"),
    col("ner.birth_date").alias("birth_date"),
    col("ner.policy_number").alias("policy_number"))

```

```python
# Save the transformed DataFrame as a new table
df_extracted.write.mode("overwrite").option("mergeSchema", "true").saveAsTable(f"{CATALOG}.{SCHEMA}.ner_extracted")
```

```python
# Show result (optional)
df_extracted.show()
```
